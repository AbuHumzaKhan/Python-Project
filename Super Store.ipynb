```python
# Filling mean in the Product Base Margin. ( We can put Mean, median, Mode.)
```

```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(),inplace = True)
```

Count of Order Priority


```python
# Replacing 'Critical ' to 'Critical'.   Becouse first critical have one space which make it unique value.
```

```python
df["Order Priority"].replace("Critical ","Critical")
```

```python
df['Order Priority'] = df["Order Priority"].replace("Critical ","Critical")
```

```python
df['Order Priority'].value_counts()
```

```python
# setting the Graph size
plt.figure(figsize = (5,4))

# Title Of Graph
plt.title('Count of Order Priority')

#Saveing the created graph
plt.savefig('Count of Order Priority.jpg')

sns.countplot(x="Order Priority", data =df)    # DF use kha se data aa rha.
# Order Priority :- kiska graph bnana hai.
# countplot :- Kya graph bnana hai.
plt.show()
```

Ship mode
